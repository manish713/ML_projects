# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

## importing the Boston House Price Dataset

```python
# importing the data and creating column names
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
house_price_dataset = pd.read_csv("/content/boston_housing.csv", header=None, delimiter=r"\s+", names=column_names)
```

```python
house_price_dataset
```

```python
# changing the target column name
house_price_dataset= house_price_dataset.rename(columns={'MEDV': 'price'})
```

```python
house_price_dataset.head()
```

```python
# checking the number of rows and columns in the data frame
house_price_dataset.shape
```

```python
# checking the missing values
house_price_dataset.isnull().sum()
```

```python
#statistical measures of dataset
house_price_dataset.describe()
```

## understanding the correlation b/w various features in the dataset

1. positive correlation
2. negative correlation

```python
correlation = house_price_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True,square = True, fmt = '.1f',annot=True,annot_kws={'size':8},cmap="Blues")
```

```python
# splitting the data and target
X = house_price_dataset.drop(['price'],axis = 1)
Y = house_price_dataset['price']
```

```python
print(X)
print(Y)
```

```python
# train test split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Model Training

### XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

## prediction on training data

```python
  # accuracy prediction on training data
  training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

## Model Evaluation

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print("Mean Absolute Error : ", score_2)
```

## visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

## prediction on test data

```python
# accuracy for prediction on test data

test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print("Mean Absolute Error : ", score_2)
```
